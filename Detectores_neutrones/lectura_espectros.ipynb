# Ejemplo de graficación de los espectros de altura de pulsos

```python
%matplotlib notebook
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from CNFreader.read_cnf import read_cnf_file
```

```python
# Archivos que se quieren leer
path_files = [
              os.path.join("Ejemplos", "He3.CNF"),
              os.path.join("Ejemplos", "BF3.CNF"),
             ]
```

```python
# Información que se puede obtener del archivo leido
data = read_cnf_file(path_files[0])
data
```

```python
# Lectura y graficación de los espectros
fig, ax = plt.subplots()
for file in path_files:
    data = read_cnf_file(file);
    cuentas = data["Channels data"]
    canales = data["Channels"]
    tiempo_vivo = data["Live time"]
    ax.plot(canales, cuentas / tiempo_vivo, label=file)
ax.set_xlabel("Canales")
ax.set_ylabel("Cuentas [cps por canal]")
ax.legend()
ax.set_title("Espectro de altura de pulsos");
```
